```python
import cv2
import numpy as np
import os
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense

import tensorflow.compat.v1 as tf


from keras.models import model_from_json
import operator
import sys
```

```python
import cv2
import numpy as np
import math

cap = cv2.VideoCapture(0)
kernel = np.ones((3, 3), np.uint8)
lower_skin = np.array([0, 20, 70], dtype=np.uint8)
upper_skin = np.array([20, 150, 255], dtype=np.uint8)

       
while (1):
    try:
        ret, frame = cap.read()
        roi_steering_wheel = frame[250:frame.shape[0], 0:frame.shape[1]]
        frame = cv2.flip(frame, 1)
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        roi = frame[250:frame.shape[0], 250:frame.shape[1]-100]

        cv2.rectangle(frame, (250, 250), (frame.shape[1]-100,frame.shape[0]), (0, 255, 0), 0)
        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
        hsv_data = cv2.cvtColor(roi_steering_wheel,cv2.COLOR_BGR2HSV)
        
        mask = cv2.inRange(hsv, lower_skin, upper_skin)
        mask_data = cv2.inRange(hsv_data, lower_skin, upper_skin)
        mask = cv2.dilate(mask, kernel, iterations=4)
        mask_data = cv2.dilate(mask_data, kernel, iterations=4)
        mask = cv2.GaussianBlur(mask, (5, 5), 100)
        mask_data = cv2.GaussianBlur(mask_data, (5, 5), 100)
        contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        cnt = max(contours, key=lambda x: cv2.contourArea(x))

        epsilon = 0.0005 * cv2.arcLength(cnt, True)
        approx = cv2.approxPolyDP(cnt, epsilon, True)

        hull = cv2.convexHull(cnt)

        areahull = cv2.contourArea(hull)
        areacnt = cv2.contourArea(cnt)

        arearatio = ((areahull - areacnt) / areacnt) * 100

        hull = cv2.convexHull(approx, returnPoints=False)
        defects = cv2.convexityDefects(approx, hull)

        l = 0

        for i in range(defects.shape[0]):
            s, e, f, d = defects[i, 0]
            start = tuple(approx[s][0])
            end = tuple(approx[e][0])
            far = tuple(approx[f][0])
            pt = (100, 180)


            a = math.sqrt((end[0] - start[0]) ** 2 + (end[1] - start[1]) ** 2)
            b = math.sqrt((far[0] - start[0]) ** 2 + (far[1] - start[1]) ** 2)
            c = math.sqrt((end[0] - far[0]) ** 2 + (end[1] - far[1]) ** 2)
            s = (a + b + c) / 2
            ar = math.sqrt(s * (s - a) * (s - b) * (s - c))

            d = (2 * ar) / a

            angle = math.acos((b ** 2 + c ** 2 - a ** 2) / (2 * b * c)) * 57

            if angle <= 90 and d > 30:
                l += 1
                cv2.circle(roi, far, 3, [255, 0, 0], -1)

            cv2.line(roi, start, end, [0, 255, 0], 2)

        l += 1
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.imshow('mask data',mask_data)
        if l == 1:
            if areacnt < 2000:
                cv2.putText(frame, 'Put hand in the box', (0, 250), font, 2, (0, 0, 255), 3, cv2.LINE_AA)
            else:
                if arearatio < 12:
                    if end[1] < 120:
                        cv2.putText(frame, 'left', (0, 250), font, 2, (0, 0, 255), 3, cv2.LINE_AA)
                        cv2.imwrite("train/left/" + "left14-"+str(i) + ".jpg", mask_data)
                        #print(mask_data.shape)
                    else:
                        cv2.putText(frame, 'right', (0, 250), font, 2, (0, 0, 255), 3, cv2.LINE_AA)
                        cv2.imwrite("train/right/" + "right14-"+str(i) + ".jpg", mask_data)

        elif l > 2 :
            cv2.putText(frame, 'STOP', (0, 250), font, 2, (0, 0, 255), 3, cv2.LINE_AA)
            cv2.imwrite("train/stop/" + "stop14-" + str(i) + ".jpg", mask_data)
        cv2.imshow('mask', mask)
        cv2.imshow('frame', frame)
    except:
        a=10

    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()

```

```python
model = Sequential()
model.add(Convolution2D(32, (3, 3), input_shape=(230, 640, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Convolution2D(32, (3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(units=128, activation='sigmoid'))
model.add(Dense(units=3, activation='softmax')) 

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
        rescale=1./255, #normalization here
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=False)

test_datagen = ImageDataGenerator(rescale=1./255) #normalization here

training_set = train_datagen.flow_from_directory('train',
                                                 target_size=(230, 640),
                                                 batch_size=5,
                                                 color_mode='grayscale',
                                                 class_mode='categorical')

test_set = test_datagen.flow_from_directory('test',
                                            target_size=(230, 640),
                                            batch_size=5,
                                            color_mode='grayscale',
                                            class_mode='categorical') 
test_set.labels
```

```python
# Loading the model
json_file = open("model-bw.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
# load weights into new model
model.save('model-bw.h5')
model.load_weights("model-bw.h5")
print("Loaded model from disk")
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 

model.fit_generator(
        training_set,
        steps_per_epoch=400,
        epochs=5,
        validation_data=test_set,
        validation_steps=20)
```

```python
# Saving the model
model_json = model.to_json()
with open("model-bw.json", "w") as json_file:
    json_file.write(model_json)
model.save_weights('model-bw.h5')
```
